```python
import pandas as pd
import matplotlib.pyplot as plt
from utils import make_subplots
```

```python
df = pd.read_csv("../processed_data/analysis_data/combined_new.csv")
df.date_minsk = pd.to_datetime(df.date_minsk)
```
