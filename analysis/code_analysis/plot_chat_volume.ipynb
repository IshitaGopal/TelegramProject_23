```python
import pandas as pd
import numpy as np
from math import ceil
import matplotlib.pyplot as plt
from utils import make_subplots
```

```python
# Read in daily volume data at the chat and district level 
chat_daily_num_msgs = pd.read_csv("../processed_data/analysis_data/chat_daily_num_msgs.csv")
chat_daily_num_msgs.date_minsk = pd.to_datetime(chat_daily_num_msgs.date_minsk)

district_daily_num_msgs = pd.read_csv("../processed_data/analysis_data/district_daily_num_msgs.csv")
district_daily_num_msgs.date_minsk = pd.to_datetime(district_daily_num_msgs.date_minsk)
```

```python
chat_daily_num_msgs.head()
```

```python
N=60
make_subplots(num_subplots=N, ncols=3, nrows=20, figsize=(3.5 * 4, 3.5 * ceil(N/4)),
              df=chat_daily_num_msgs, grouping_var='chat_labels', x='date_minsk', y='num_msgs', logy=True, 
              title="Daily Log(Volume of Messages) across 59 group chats (August 9th - 1st April)")
```
